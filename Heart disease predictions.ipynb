```python
# finding the optimum depth of the tree
depth = np.arange(1,20)
s = []

for k in depth:
    dec_tree = DecisionTreeClassifier(max_depth=k)
    dec_tree.fit(X_train,y_train)
    score = recall_score(y_valid, dec_tree.predict(X_valid), average='binary')
    s.append(score)
```

```python
plt.plot(depth , np.array(s))
plt.show()
```

```python
leaf = np.arange(1,10)
s = []

for k in leaf:
    dec_tree = DecisionTreeClassifier(min_samples_leaf=k)
    dec_tree.fit(X_train,y_train)
    score = recall_score(y_valid, dec_tree.predict(X_valid), average='binary')
    s.append(score)
```

```python
plt.plot(leaf , np.array(s))
plt.show()
```
